## Notation
**Set notation**
- A **set** is an unordered collection of 'things' denoted by curly brackets $\{$,$\}$. For example, if we want to define a set called $S$ containing the four numbers $1,2,3,4$ (in any particular order), we write $S=\{1,2,3,4\}$. Sets are not limited to containing numbers but can be defined to include any 'things' we can imagine: vectors, functions, matrices, problem solutions, apples, bananas... Mathematical sets are not that different than sets in python. However, in mathematics we tend to make sure that sets only contain 'things' of the same 'type'.
- Sets can also be infinite. For example, the set of real numbers is denoted $\mathbb{R}$, the set of natural numbers $\mathbb{N} = \{1,2,3,...\}$ sometimes defined to include 0), the set of integers $\mathbb{Z} = \{...,-2,-1,0,1,2,...\}$.
- If a 'thing' $x$ belongs to a set $S$, we write $x \in S$, reads as "$x$ is in the set S". For example, if a number $a$ is a real number, we write $a \in \mathbb{R}$. 
- Sets notation can also define sets as selections of larger sets, called set-builder notation. For example,
    - The set of rational numbers (numbers that can be written as fractions of integers) can be defined as $\mathbb{Q} = \{ \frac{p}{q} : p,q\in\mathbb{Z}, q\neq 0 \}$, reads as "$\mathbb{Q}$ is the set of numbers $\frac{p}{q}$ **such that** $p$ and $q$ are integers and $q$ is not zero".
    - The set of $n$-dimmensional vectors $v$ having length 1 can be defined as $\{ v\in\mathbb{R}^n : \mathrm{length}(v) = 1 \}$, where $\mathrm{length}$ is some function defining a length.

# Ch 5. Solving linear least-squares problems

## Fitting a line to data
Consider the simplest possible regression problem of fitting a line to a set of points. The equation of a line
\begin{equation}
\hat{y}_k = a_1x_k + a_0,
\end{equation}
where $a_0,a_1$ are unknown parameters to be determined from a collection of $(x_i,y_i)$ pairs. Given $N-1$ input-output pairs $(x_i,y_i)$, $i=0,1,...,N-1$, we define the vectors
\begin{equation}
x = \begin{bmatrix}
x_0\\x_1\\x_2\\ \vdots \\ x_{N-1}
\end{bmatrix},
y = \begin{bmatrix}
y_0\\y_1\\y_2\\ \vdots \\ y_{N-1}
\end{bmatrix}.
\end{equation}
Defining the parameter vector and feature vector
\begin{equation}
\theta =
\begin{bmatrix}
a_0 \\ a_1
\end{bmatrix},
~\varphi(x_k) = \begin{bmatrix}
1\\x_k
\end{bmatrix}
\end{equation}
respectively, the predicted output is
\begin{equation}
\hat{y}_k = \theta^T\varphi(x_k).
\end{equation}
Fitting the line to the set of data points i by minimizing the least-squares cost is equivalent to finding the $\theta$ that minimizes the cost function
\begin{equation}
J(\theta) = \sum_{k=0}^{N-1} ( y_k - \hat{y}_k )^2 =  \| y - \hat{y} \|_2^2.
\end{equation}

Defining the matrices
\begin{equation}
X = 
\begin{bmatrix}
1& x_0\\
1& x_1\\
\vdots &\vdots\\
1& x_{N-1}\\
\end{bmatrix},
\end{equation}
we have that $\hat{y} = X\theta$, and the cost function can be written in terms of matrix-vector multiplication as
\begin{equation}
J(\theta) = (y-X\theta)^T(y-X\theta).
\end{equation}
Minimum of $J(\theta)$ is obtained when
\begin{align}
\frac{\mathrm{d}J}{d\theta} = -2X^Ty  + 2X^TX\theta = 0\\
\implies~ \hat{\theta} = (X^TX)^{-1}X^Ty.
\end{align}

The following example illustrates the linear least-squares problem.

```python
N = 10 #number of data points
x = np.linspace(0,1,num=N) 
np.random.seed(0)
v = np.random.normal(0,0.1,N) #measurement noise
y = 2*x + 1 + v #measured output

# -----------------------------------------------
def lin_regressor(x):
    '''
    creates a regressor matrix X for linear regression for the model
    yhat_k = a1*x_k + a_0, theta = [a_0, a_1]^T,
    such that
    y = X*theta
    '''
    X = []
    for xi in x:
        X.append([1,xi])
    X = np.array(X)
    return X

def ls_solve(X,y):
    '''Solves a linear least squares problem for regressor X and output y'''
    theta = np.linalg.inv(X.T@X)@X.T@y
    return theta
# -----------------------------------------------

#solve linear least squares problem
X = lin_regressor(x)
theta_hat = ls_solve(X,y)

# -----------------------------------------------
#plot results
x_hat = np.linspace(0,1,num=100)
X = lin_regressor(x_hat)
y_hat = X@theta_hat

plt.scatter(x,y,color='blue')
plt.plot(x_hat,y_hat,color='red')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
print(theta_hat)
```

## Fitting a polynomial to data
Next consider fitting a $n$th order polynomial 
\begin{equation}
\hat{y}_k = a_0 + a_1x_k + a_2x_k^2 + ... + a_nx^n
\end{equation}
to data points. This can be done exactly in the same way as fitting a line by defining the regressor and variables as
\begin{equation}
X =
\begin{bmatrix}
1 &x_0 &x_0^2 &x_0^3 &\cdots &x_0^n\\
1 &x_1 &x_1^2 &x_1^3 &\cdots &x_1^n\\
\vdots &\vdots &\vdots &\vdots & &\vdots\\
1 &x_{N-1} &x_{N-1}^2 &x_{N-1}^3 &\cdots &x_{N-1}^n\\
\end{bmatrix},~
\theta =
\begin{bmatrix}
a_0\\
a_1\\
a_2\\
\vdots\\
a_n
\end{bmatrix}.
\end{equation}

The following example illustrates fitting i polynomial to data points using least-squares regression. Note that as the polynomial order increases, the fit to the data points used for training improves, but the generalization (fit in between data points) might get poor!

```python
#data
theta_true = np.array([0.5, 5, -15, 10]) #ground truth
y = theta_true[3]*np.power(x,3) + theta_true[2]*np.power(x,2) + theta_true[1]*x + theta_true[0] + v

# -----------------------------------------------
POLYNOMIAL_ORDER = 8 #order of the polynomial
# -----------------------------------------------

def poly_regressor(x,POLYNOMIAL_ORDER):
    '''
    creates a regressor matrix X for polynomial regression for the model
    yhat_k = a0 + a1*x_k + a_2x_k^2 + ..., theta = [a_0, a_1, a_2,...]^T,
    such that
    y = X*theta
    '''
    X = []
    for xi in x:
        phi=[]
        for p in range(POLYNOMIAL_ORDER+1):
            phi.append(np.power(xi,p))
        X.append(phi)
    X = np.array(X)
    return X

# -----------------------------------------------
# solve linear least-squares problem
X = poly_regressor(x,POLYNOMIAL_ORDER)
theta_hat = ls_solve(X,y)

# -----------------------------------------------
#plot results
X = poly_regressor(x_hat,POLYNOMIAL_ORDER)
X_true = poly_regressor(x_hat,3)
y_hat = X@theta_hat
y_true = X_true@theta_true

plt.scatter(x,y,color='blue',label='data')
plt.plot(x_hat,y_hat,color='red',label='fit')
plt.plot(x_hat,y_true,'k--',label='ground truth')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('order='+str(POLYNOMIAL_ORDER))
plt.legend()
plt.show()
print(theta_hat)
```

# Ch.6 Regularized least squares
Regularization is often applied to remedy the problem of overfitting. Instead of solving a linear least-squares problem, a penalty function $L(\theta)$ constraining the 'size' of $\theta$ is added to the cost function. Resulting in the cost function
\begin{equation}
J(\theta) = (y-X\theta)^T(y-X\theta) + L(\theta).
\end{equation}
Determining what the regularization term is is an engineering problem and is the task for the user to figure out for a given application. Below are a few common examples of regularized least squares problems.

## Ridge regression (Tokhonov regularization, $\ell_2$ regularization)

Ridge regression is a technique used in regression analysis to prevent overfitting by adding a penalty term to the traditional least squares objective function. This penalty term shrinks the coefficients towards zero, effectively reducing their variance and stabilizing the model.. Ridge regressionis likely the simplest and most commonly used type of regularization for regression. The regularization term is
\begin{equation}
L(\theta) = \lambda \theta^T\theta,
\end{equation}
where $\lambda>0$ is a so-called regularization parameter determining the tradeoff between goodness of fit and size of $\theta$. For small $\lambda$ the solution is close to that of standard least-squares, for a very large $\lambda$ all variables $\theta$ tend to zero.

The ridge-regression problem minimizes the cost function
\begin{equation}
J(\theta) = (y-X\theta)^T(y-X\theta) + \lambda \theta^T\theta =  \| y - \hat{y} \|_2^2 + \lambda\|\theta\|_2^2,
\end{equation}
and regularized solution $\hat{\theta}$ is obtained from  from the optimality criteria
\begin{equation}
\frac{\mathrm{d}J(\theta)}{\mathrm{d}\theta}=0, \implies~ \hat{\theta} = (X^TX + \lambda I)^{-1}X^Ty
\end{equation}

The example below illustrates how regularization can be used to prevent overfitting in the polynomial-regression problem above...

```python
def ridge_solve(X,y,lam):
    '''Solves a linear least squares problem for regressor X and output y'''
    N = y.size
    theta = np.linalg.inv(X.T@X + lam*np.eye(POLYNOMIAL_ORDER+1))@X.T@y
    return theta
    
# -----------------------------------------------
POLYNOMIAL_ORDER = 8 #order of the polynomial
LAM = 0.001 #regularization parameter
# -----------------------------------------------

# fit polynomial
X = poly_regressor(x,POLYNOMIAL_ORDER)
theta_hat = ridge_solve(X,y,LAM)

# -----------------------------------------------
#plot results
X = poly_regressor(x_hat,POLYNOMIAL_ORDER)
X_true = poly_regressor(x_hat,3)
y_hat = X@theta_hat
y_true = X_true@theta_true

plt.scatter(x,y,color='blue')
plt.plot(x_hat,y_hat,color='red')
plt.plot(x_hat,y_true,'k--')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('order='+str(POLYNOMIAL_ORDER))
plt.show()
print(theta_hat)
```

## Hodrick-Prescott trend filter
The Hodrick-Prescott (H-P) trend filter aims to approximate noisy $y$ with a smooth signal $\hat{y}$ by constraining the second difference $\hat{y}(k) = y(k-1) - 2y(k) + y(k+1)$. This is obtaind by constraining the second difference of $\hat{y}$. The H-P filter minimizes the cost function
\begin{equation}
J(\hat{y}) = (y-\hat{y})^T(y-\hat{y}) + \lambda \hat{y}^T D^T D \hat{y} = \|y-\hat{y}\|_2^2 + \lambda\|D\hat{y}\|_2^2,
\end{equation}
where $D$ is the second difference matrix
\begin{equation}
D = \begin{bmatrix}
1 &-2 &1 &0 &\cdots &0\\
0 &1 &-2 &1 &\cdots &0\\
  & &\ddots &\ddots &\ddots.
\end{bmatrix}
\end{equation}
The regularization term $\lambda\|D\hat{y}\|_2^2$ enforces the signal $\hat{y}$ to be smooth. The solution to the H-P trend filter is a weighted, tikhonov-regularized least-squares problem which has the solution
\begin{equation}
\hat{y} = (I + \lambda D^T D)^{-1}y.
\end{equation}

*Note:* The H-P filter is not a regression method, but a pre-processing step used to revieal ternds in noisy data.

```python
# create data
N = 100 #number of data points
x = np.linspace(0,1,num=N)
np.random.seed(0)
v = np.random.normal(0,0.1,N) #measurement noise
y = theta_true[3]*np.power(x,3) + theta_true[2]*np.power(x,2) + theta_true[1]*x + theta_true[0] + v

# -----------------------------------------------
LAMBDA = 500
# -----------------------------------------------

def diff_diff_mat(N):
    D = np.diag(np.ones(N), k=0) - 2*np.diag(np.ones(N-1), k=1) + np.diag(np.ones(N-2), k=2)
    D = D[:-2]
    return D

def hpfilter(y,lam):
    '''Hodrick-Prescott (H-P) trend filtering approzximates a signal y with a smooth signal y_hat
    by constraining the second difference.'''
    N = y.size
    D = diff_diff_mat(N)
    yhat = np.linalg.inv(np.eye(N) + lam*D.T@D)@y
    return yhat
    
# -----------------------------------------------
# H-P filter
y_hat = hpfilter(y,LAMBDA)

# -----------------------------------------------
# Plot results
plt.scatter(x,y,color='blue')
plt.plot(x,y_hat,'o--',color='red')
plt.plot(x_hat,y_true,'k--')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$\lambda$ = '+str(LAMBDA))
plt.show()
```

## *Least-Squares Support Vector Machines (LS-SVM)
Continuing the example of the linear regression, as seen in the example above, the number of parameters $\theta_i$ to estimate increases with the number of basis functions used to represent the input. Consider the, seemingly redicilous, example of representing $x$ in terms of an infinite sequence $\varphi_0(x_k),\varphi_1(x_k),...$, resulting in
\begin{equation}
\varphi(x_k) = 
\begin{bmatrix}
\varphi_0(x_k)\\
\varphi_1(x_k)\\
\varphi_2(x_k)\\
\vdots
\end{bmatrix}
\quad\text{and}\quad
y_k = \theta_0 + \sum_{i=1}^\infty \theta_i \varphi_i(x_k) = x_0 + X\theta,
\end{equation}
where
\begin{equation}
X =
\begin{bmatrix}
\varphi_0(x_0)^T &\varphi_1(x_0) &\cdots\\
\varphi_0(x_1)^T &\varphi_1(x_1) &\cdots\\
\vdots &\vdots
\end{bmatrix}
\end{equation}
Ignoring the fact that we have infinitely many unknowns $\theta_i$ and that we would never be able to collect enough data to fit the parameters, we can formulate the problem as a constrained least-squares problem
\begin{align*}
&\underset{e,\theta,\theta_0}{\text{minimize}}\quad \frac{1}{2}e^Te + \frac{\gamma}{2} \theta^T\theta\\
&\text{subject to} \\
&e_k = y_k - \theta^T\varphi(x_k)-\theta_0,~k=0,1,...,N,
\end{align*}
$\gamma>0$. Incorporating the equality constraints into the objective function by introducing [Lagrange multipliers](https://en.wikipedia.org/wiki/Lagrange_multiplier) $\lambda_i$, we get the Lagrangian
\begin{equation}
\mathcal{L}(\theta_0,\theta,\alpha,e) = \frac{1}{2}\theta^T\theta + \frac{\gamma}{2}e^Te - \sum_{k=0}^{N-1}\lambda_k( \theta^T\varphi(x_k) + \theta_0 + e_k - y_k ),
\end{equation}
which is essentially the same problem as the regularized least-square but with $N$ number of unknown Lagrange multipliers. The optimal solution to the original problem is obtained from the Lagrangian when the following conditions are satisfied
\begin{align}
\frac{\partial\mathcal{L}}{\partial\theta} &= 0 ~\implies~ \theta = \varphi(x)\lambda,\\
\frac{\partial\mathcal{L}}{\partial\theta_0} &= 0 ~\implies~ \sum_{k=0}^{N-1} \lambda_k = 0,\\
\frac{\partial\mathcal{L}}{\partial e} &= 0 ~\implies~ \lambda = \gamma e \\
\frac{\partial\mathcal{L}}{\partial\lambda} &= 0 ~\implies~ \varphi(x)^T\theta + \theta_0 + e - y = 0. 
\end{align}
Note the suprising fact that the infinite number of original unknowns $\theta_i$ can be eliminated by backwards substitution, $e=\gamma^{-1}\lambda$, resulting in
\begin{align}
y &= \theta_0 + (\varphi(x)^T\varphi(x) + \gamma^{-1}I)\lambda\\
0 &= 1_N^T\lambda, 
\end{align}
which can be written as

\begin{equation}
\begin{bmatrix}
0 &1_N^T \\
1_N & \varphi(x)^T\varphi(x) + \gamma^{-1}\lambda
\end{bmatrix}
\begin{bmatrix}
\theta_0 \\
\lambda
\end{bmatrix}
=
\begin{bmatrix}
0\\y
\end{bmatrix}.
\end{equation}
The mind-blowing revelation is that this system only has a finite number of unknowns ($N$), the same number of observations, and the inner product $\varphi(x)^T\varphi(x)$ is an $N$-by-$N$ matrix.

The inner product is called a Kernel, and is often denoted
\begin{equation}
K(x_k,x_j) = \varphi(x_k)^T\varphi(x_j),
\end{equation}
and as long as we can select the basis such that the Kernel function can be easily calculated, we can solve this regression problem with an infinite number of unknowns!
