```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate some sample data (for example purposes)
np.random.seed(42)
time_series = np.cumsum(np.random.randn(100)) + np.linspace(0, 10, 100)

# Apply HP filter (using 1600 for quarterly data as an example for lambda)
cycle, trend = sm.tsa.filters.hpfilter(time_series, lamb=1600)

# Plotting the results
plt.figure(figsize=(14, 7))
plt.plot(time_series, label='Original Data')
plt.plot(trend, label='Trend', linestyle='--')
plt.plot(cycle, label='Cycle', linestyle=':')
plt.legend()
plt.show()
```

```python
import numpy as np

# Sample data
Y = np.array([1, 2, 3, 4, 5])

# Smoothing parameter
lambda_hp = 100  # Just for demonstration, normally this would be chosen based on data frequency

# Constructing the D matrix for second differences
D = np.array([
    [1, -2, 1, 0, 0],
    [0, 1, -2, 1, 0],
    [0, 0, 1, -2, 1]
])

# Identity matrix
I = np.eye(5)

# Solving for T
T = np.linalg.inv(I + lambda_hp * D.T @ D) @ Y

T
```

```python
def generate_d_array(n):
    rows = n-2
    arr = []
    for i in range(rows):
        arr.append([0]*(n-1))
        arr[i][i:i+2] = [1,-2,1]

    return np.array(arr)


```
