```python
X = [95, 85, 80, 70, 60] 
Y = [90, 80, 70, 65, 60] 
n = len(X)
```

```python
def calculateB(x, y, n):
   
    # sum of array x 
    sx = sum(x)
 
    # sum of array y 
    sy = sum(y)
     
    # for sum of product of x and y 
    sxsy = 0
 
    # sum of square of x 
    sx2 = 0
 
    for i in range(n):
        sxsy += x[i] * y[i]
        sx2 += x[i] * x[i]
    b = (n * sxsy - sx * sy)/(n * sx2 - sx * sx)
    return b
 
# Function to find the 
# least regression line
def leastRegLine(X,Y,n):
     
    # Finding b 
    b = calculateB(X, Y, n)
    meanX = int(sum(X)/n)
    meanY = int(sum(Y)/n)
 
    # Calculating a
    a = meanY - b * meanX
 
    # Printing regression line 
    print("Regression line:")
    print("Y = ", '%.3f'%a, " + ", '%.3f'%b, "*X", sep="")
    return a, b
```

```python
a, b = leastRegLine(X, Y, n)
a, " ", b
```

```python
import numpy as np

def generate_random_data(min_val, max_val, num_points):
    x_points = np.random.uniform(min_val, max_val, num_points)
    y_points = np.random.uniform(min_val, max_val, num_points)
    
    return x_points, y_points
```

```python
import numpy as np

x , y  = generate_random_data(0, 100, 12)

A = np.stack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
m, c
```

```python
a = m # The slope of your best fit line
b = c # The intercept of your best fit line

# Calculate the y values of the best fit line
y_best_fit = [a * x + b for x in x]
```

```python
import matplotlib.pyplot as plt




ax1 = plt.plot(x, y )
ax1


```

```python
plt.scatter(x, y, color='blue', label='Data Points')

# Plot the best fit line
plt.plot(x, y_best_fit, color='red', label='Best Fit Line')

# Labeling the axes
plt.xlabel('X')
plt.ylabel('Y')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
